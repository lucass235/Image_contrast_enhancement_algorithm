```python
# Pegar dataset de um arquivo txt
df = pd.read_csv('./metricas_kodim/metricas_kodim24.png.txt', sep=',', header=None, names=['pSize', 'psnr', 'ssim', 'vif'])
name_image : str = 'Kodim24'
```

```python
# retirar as labels de cada coluna
df = df.applymap(lambda x: x.split(': ')[1])
# Arredondar os valores para 6 casas decimais
df = df.applymap(lambda x: round(float(x), 6))
```

```python
df.head(10)
```

```python
# plotar o gráfico
df.plot(figsize=(10, 5), grid=True, style='.-')
plt.xlabel('Pupulation Size')
plt.ylabel('Métrica')
plt.title('Métricas para Pupulation Size')
plt.show()
```

```python
# plotar o gráfico somente do PSNR

plt.plot(df['pSize'], df['psnr'], 'o-', color='blue')
plt.xlabel('pSize')
plt.ylabel('PSNR')
plt.title(name_image)
plt.xticks([5, 10, 20, 30, 50])  # Definir os valores do eixo x
plt.legend(['iEBMO'])
plt.savefig(f"./grafico_metricas/Kodim/{name_image}/psnr_{name_image}.png")
plt.show()
```

```python
# plotar grafico somente do SSIM

plt.plot(df['pSize'], df['ssim'], 'o-', color='green')
plt.xlabel('pSize')
plt.ylabel('SSIM')
plt.title(name_image)
plt.xticks([5, 10, 20, 30, 50])  # Definir os valores do eixo x
plt.legend(['iEBMO'])
plt.savefig(f'./grafico_metricas/Kodim/{name_image}/ssim_{name_image}.png')
plt.show()
```

```python
# plotar grafico somente do VIF

plt.plot(df['pSize'], df['vif'], 'o-', color='magenta')
plt.xlabel('pSize')
plt.ylabel('VIF')
plt.title(name_image)
plt.xticks([5, 10, 20, 30, 50])  # Definir os valores do eixo x
plt.legend(['iEBMO'])
plt.savefig(f'./grafico_metricas/Kodim/{name_image}/vif_{name_image}.png')
plt.show()
```
