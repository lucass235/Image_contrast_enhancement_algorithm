```python
import pandas as pd
import random
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
# pegar o caminho de todas imagens da pasta

pasta = './metricas_kodim/'  # Substitua pelo caminho correto da pasta

# Obtém uma lista de todos os arquivos na pasta
arquivos = os.listdir(pasta)

# Filtra apenas os arquivos TXT
txts = [arquivo for arquivo in arquivos if arquivo.endswith('.txt')]

# Obtém o caminho completo de cada arquivo TXT
caminhos_txts = [os.path.join(pasta, txt) for txt in txts]

# Imprime os caminhos dos arquivos TXT
for caminho_txt in caminhos_txts:
    print(caminho_txt)
```

```python
medias_imagens = []
for caminho_txt in caminhos_txts:
    df = pd.read_csv(caminho_txt, sep=',', header=None, names=['pSize', 'psnr', 'ssim', 'vif'])
    name_image = caminho_txt.split('/')[-1].split('.')[0]
    # retirar as labels de cada coluna
    df = df.applymap(lambda x: x.split(': ')[1])
    # Arredondar os valores para 6 casas decimais
    df = df.applymap(lambda x: round(float(x), 6))
    # calcular a média de cada coluna
    media_psnr = df['psnr'].mean()
    media_ssim = df['ssim'].mean()
    media_vif = df['vif'].mean()
    # print('Média PSNR: ', media_psnr)
    # print('Média SSIM: ', media_ssim)
    # print('Média VIF: ', media_vif)
    # armazenar os valores da media em um array de objetos
    media = [name_image.replace('metricas_', '') ,media_psnr, media_ssim, media_vif]
    # salvar as medias no array
    medias_imagens.append(media)	
    
# df = pd.read_csv('./metricas_kodim/metricas_kodim24.png.txt', sep=',', header=None, names=['pSize', 'psnr', 'ssim', 'vif'])
# name_image = 'kodim24'

# # retirar as labels de cada coluna
# df = df.applymap(lambda x: x.split(': ')[1])
# # Arredondar os valores para 6 casas decimais
# df = df.applymap(lambda x: round(float(x), 6))

# df.head()

# # calcular a média de cada coluna
# media_psnr = df['psnr'].mean()
# media_ssim = df['ssim'].mean()
# media_vif = df['vif'].mean()
# print('Média PSNR: ', media_psnr)
# print('Média SSIM: ', media_ssim)
# print('Média VIF: ', media_vif)

# # armazenar os valores da media em um array de objetos
# media = [media_psnr, media_ssim, media_vif]
```
